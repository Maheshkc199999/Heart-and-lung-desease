```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
```

```python
X= df.drop('LUNG_DESEASE', axis = 1)
y= df['LUNG_DESEASE']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
X_train_scaled=pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled=pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression(random_state=0)
lr_model.fit(X_train, y_train)
```

```python
lr_pred= lr_model.predict(X_test)
lr_pred
```

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score
lr_cr=classification_report(y_test, lr_pred)
print(lr_cr)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model= DecisionTreeClassifier(criterion='entropy', random_state=0)  
dt_model.fit(X_train, y_train)
```

```python
dt_pred= dt_model.predict(X_test)
dt_pred
```

```python
dt_cr=classification_report(y_test, dt_pred)
print(dt_cr)
```

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(X_train, y_train)
```

```python
svc_pred= svc_model.predict(X_test)
svc_pred
```

```python
svc_cr=classification_report(y_test, svc_pred)
print(svc_cr)
```

```python
from sklearn.neighbors import KNeighborsClassifier  
knn_model= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )  
knn_model.fit(X_train, y_train)
```

```python
knn_pred= knn_model.predict(X_test)
knn_pred
```

```python
knn_cr=classification_report(y_test, knn_pred)
print(knn_cr)
```

```python
from xgboost import XGBClassifier
xgb_model = XGBClassifier()
xgb_model.fit(X_train, y_train)
```

```python
xgb_pred= xgb_model.predict(X_test)
xgb_pred
```

```python
xgb_cr=classification_report(y_test, xgb_pred)
print(xgb_cr)
```
